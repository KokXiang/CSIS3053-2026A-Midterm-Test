### Q4. A charity organisation (CompanyA) uses AI to predict food demand. Assume Hari Raya is coming, CompanyA plans to give out lunchboxes to 40 families for 30 days.

#### (a) Explain the problem and why AI is suitable. (6 marks)

For a charity organization like CompanyA, accurately predicting food demand for events like Hari Raya is crucial. The core problem lies in balancing the supply of lunchboxes with the actual demand from 40 families over 30 days.

1.  **Risk of Wastage:** Overestimation of demand leads to excess food, which can result in spoilage, waste of resources (ingredients, preparation time, volunteer effort, money), and environmental impact.
2.  **Risk of Unmet Needs:** Underestimation of demand means some families might not receive enough food, leading to hunger, dissatisfaction, and a failure to meet the charity's mission.
3.  **Logistical Challenges:** Inaccurate predictions complicate logistics, such as procurement of ingredients, scheduling volunteers, cooking capacity, and distribution planning.
4.  **Variability:** Demand can fluctuate due to various factors like changes in family composition, last-minute dietary needs, participation rates, and external events.

**Why AI is suitable:**

AI, particularly machine learning, is exceptionally well-suited to address this problem due to its ability to process complex data and identify patterns that are difficult for humans to discern.

1.  **Pattern Recognition and Predictive Accuracy:** AI models can analyze historical data (e.g., past distribution records, seasonal demand, holiday impact, demographic information) to identify subtle patterns and trends. This allows for more accurate forecasting of future demand compared to manual estimations or simple statistical methods.
2.  **Handling Multiple Variables:** Food demand is influenced by numerous factors (e.g., number of families, average family size, specific holiday period, socio-economic conditions, even weather). AI can integrate and weigh these diverse variables simultaneously to provide a comprehensive prediction.
3.  **Adaptability and Continuous Improvement:** As new data becomes available (e.g., actual demand during previous Hari Raya periods, feedback from families), AI models can be continuously retrained and updated. This allows the system to learn from its mistakes, adapt to changing circumstances, and improve its predictive performance over time.
4.  **Optimization of Resources:** By providing precise demand predictions, AI helps CompanyA optimize its resource allocation, minimizing food waste and ensuring that every lunchbox prepared directly meets a need. This leads to greater operational efficiency and financial sustainability for the charity.
5.  **Proactive Decision-Making:** With reliable forecasts, CompanyA can make proactive decisions regarding purchasing, volunteer scheduling, and logistical arrangements, reducing last-minute stress and improving overall service delivery.

A complete AI solution for predicting food demand involves several key stages, from data collection to deployment and continuous monitoring. Below is a conceptual overview, followed by a Python demonstration of how an AI system could 'calculate' food demand and cost for the upcoming 30 days.

**Conceptual AI Solution Overview:**

**1. Problem Definition & Objective:**
*   **Objective:** To accurately predict the daily number and composition of lunchboxes required for 40 families over 30 days during Hari Raya, minimizing both surplus and shortage.
*   **Key Performance Indicators (KPIs):** Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), or Mean Absolute Percentage Error (MAPE) for prediction accuracy; percentage of food waste; percentage of unmet demand.

**2. Data Collection:**
*   **Historical Demand Data:** Records of past lunchbox distributions, including dates, number of lunchboxes distributed, number of beneficiaries, types of food items, and any recorded wastage or shortages.
*   **Beneficiary Demographics:** Data on the 40 families, such as average family size, age distribution within families (e.g., number of children vs. adults), any recorded dietary restrictions (e.g., vegetarian, allergies), and preferences (if available).
*   **Calendar/Event Data:** Dates of public holidays (Hari Raya), school holidays, local community events, or any other factors that might influence family presence or demand.
*   **External Factors (Optional but beneficial):** Local economic indicators, weather patterns (e.g., extreme heat affecting appetite), or other relevant socio-economic data.

**3. Data Preprocessing:**
*   **Data Cleaning:** Handle missing values (e.g., imputation or removal), correct inconsistencies (e.g., typos in food names), and identify/remove outliers.
*   **Feature Engineering:** Create new features that could be predictive:
    *   `DayOfWeek`: Extract day of the week from dates.
    *   `IsHoliday`: Binary flag for Hari Raya or other significant holidays.
    *   `DaysUntilHariRaya`: Number of days leading up to or after the main Hari Raya date.
    *   `FamilySizeGroup`: Categorize families by size (e.g., small, medium, large).
    *   `SeasonalIndicators`: Features like `month`, `quarter`.
*   **Encoding Categorical Variables:** Convert categorical features (e.g., dietary restrictions, food types) into numerical representations using techniques like One-Hot Encoding.
*   **Data Aggregation:** Aggregate historical data to the desired prediction granularity (e.g., daily lunchbox count).
*   **Splitting Data:** Divide the preprocessed dataset into training, validation, and test sets (e.g., 70% train, 15% validation, 15% test), ensuring chronological order is maintained for time-series data.

**4. Model Selection:**
*   **Time Series Models:** For predicting daily lunchbox counts over 30 days, models like ARIMA, SARIMA, Prophet (Facebook's time series model), or more advanced deep learning models like LSTMs (if sufficient historical data is available) are suitable.
*   **Regression Models:** If treating each day's prediction independently based on features, models like Random Forest Regressor, Gradient Boosting Regressor (e.g., XGBoost, LightGBM), or a simple Linear Regression could be considered as baselines.
*   **Ensemble Methods:** Combining predictions from multiple models can often lead to improved accuracy.

**5. Model Training & Validation:**
*   **Training:** Train the selected model(s) on the training dataset. The model learns the relationships between the input features and the target variable (number of lunchboxes).
*   **Hyperparameter Tuning:** Optimize model parameters (e.g., number of trees in a Random Forest, learning rate in Gradient Boosting) using techniques like GridSearchCV, RandomizedSearchCV, or Bayesian Optimization on the validation set.
*   **Cross-Validation:** Implement time-series cross-validation (e.g., rolling origin or expanding window) to ensure the model's robustness and generalization ability over different time periods.

**6. Model Evaluation:**
*   **Testing:** Evaluate the final trained and tuned model on the unseen test dataset.
*   **Metrics:** Calculate KPIs like MAE, RMSE, MAPE to quantify prediction accuracy. Analyze errors to understand where the model performs well and where it struggles.
*   **Bias Analysis:** Check for any systematic under- or over-prediction and analyze if it affects specific groups or time periods.

**7. Deployment & Monitoring:**
*   **Deployment:** Integrate the trained model into an operational system. This could be a simple script that runs daily to generate predictions, or a more sophisticated web application/dashboard.
*   **Prediction Generation:** The system will take current input features (e.g., 'today's date', 'is Hari Raya', 'family info') and generate predictions for the next 30 days.
*   **Integration with Operations:** The predictions are then provided to CompanyA's operational team for procurement, preparation, and distribution planning.
*   **Continuous Monitoring:** Regularly monitor the model's performance in a real-world setting. Compare actual demand with predicted demand to track accuracy over time.
*   **Retraining Strategy:** Establish a schedule for retraining the model with new, updated data (e.g., monthly, quarterly, or before major events like Hari Raya) to ensure it remains accurate and adapts to new patterns or changes in circumstances.

#### 3. AI Algorithm for Lunchbox Demand Prediction

We'll re-use and enhance the linear regression model for predicting lunchbox demand. We will consider the `average_ingredient_cost_per_day` as a potential feature, as food prices might indirectly influence logistical planning or available budget for lunchboxes, although for fixed charity distribution, it might primarily impact budgeting rather than the number of boxes directly requested by families. For this model, we will keep it simple and focus on day-of-week and holiday effects first, then combine with cost later.

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import datetime

# --- 1. Generate Synthetic Historical Demand Data (similar to previous approach) ---
# This will represent the demand over a year
historical_demand_dates = pd.date_range(start='2026-01-01', periods=365, freq='D')
historical_df = pd.DataFrame({'date': historical_demand_dates})
historical_df['day_of_week'] = historical_df['date'].dt.dayofweek
historical_df['is_weekend'] = ((historical_df['date'].dt.dayofweek == 5) | (historical_df['date'].dt.dayofweek == 6)).astype(int)

# Simulate a mock holiday period (e.g., around late year)
mock_holiday_start = pd.to_datetime('2026-12-01')
mock_holiday_end = pd.to_datetime('2026-12-15')
historical_df['is_holiday'] = ((historical_df['date'] >= mock_holiday_start) & (historical_df['date'] <= mock_holiday_end)).astype(int)

# Base demand for 40 families (e.g., 1 lunchbox per family, some families might need 2, some might not be there)
# Let's assume average 1.5 lunchboxes per family * 40 families = 60 base lunchboxes
base_demand = 60

# Simulate demand with weekly patterns, holiday bumps, and noise
historical_df['demand'] = base_demand \
                         + historical_df['day_of_week'].apply(lambda x: -5 if x in [0, 1] else (5 if x in [4, 5] else 0)) \
                         + historical_df['is_holiday'] * 15 \
                         + np.random.randint(-5, 5, size=len(historical_df))

# Ensure demand is non-negative and realistic for 40 families (e.g., 40-80 range)
historical_df['demand'] = historical_df['demand'].apply(lambda x: max(x, 40)) # Min demand 40 (1 per family)
historical_df['demand'] = historical_df['demand'].apply(lambda x: min(x, 80)) # Max demand 80 (2 per family)

# Merge historical demand with average ingredient cost for potential feature inclusion
historical_df = pd.merge(historical_df, food_prices_df[['date', 'average_ingredient_cost_per_day']], on='date', how='left')
historical_df['average_ingredient_cost_per_day'] = historical_df['average_ingredient_cost_per_day'].fillna(historical_df['average_ingredient_cost_per_day'].mean()) # Fill any NaNs if dates don't perfectly align


# --- 2. Prepare Data for Training ---
features = ['day_of_week', 'is_weekend', 'is_holiday', 'average_ingredient_cost_per_day'] # Added cost as a feature
target = 'demand'

X = historical_df[features]
y = historical_df[target]

# --- 3. Train a Simple Predictive Model ---
model = LinearRegression()
model.fit(X, y)

print("Model trained successfully.")

# --- 4. Generate Future Data for Prediction (Hari Raya Period) ---
# Assume Hari Raya is from 2026-04-10 for 30 days
hari_raya_start = pd.to_datetime('2026-03-21') # Updated Hari Raya start date
future_dates = pd.date_range(start=hari_raya_start, periods=30, freq='D')
future_df = pd.DataFrame({'date': future_dates})
future_df['day_of_week'] = future_df['date'].dt.dayofweek
future_df['is_weekend'] = ((future_df['date'].dt.dayofweek == 5) | (future_df['date'].dt.dayofweek == 6)).astype(int)

# 'is_hari_raya' flag for the entire 30-day period
future_df['is_holiday'] = 1 # Mark entire 30 days as 'holiday' for Hari Raya effect

# For future ingredient costs, we'll assume a continuation of the last known average cost or a slight trend
# In a real scenario, you'd use forecasted food prices
last_avg_cost = food_prices_df['average_ingredient_cost_per_day'].iloc[-1]
future_df['average_ingredient_cost_per_day'] = last_avg_cost + np.random.normal(0, 0.1, size=len(future_df)) # Simulate slight fluctuation
future_df['average_ingredient_cost_per_day'] = future_df['average_ingredient_cost_per_day'].apply(lambda x: max(x, 1.0))

future_features = future_df[features]

# --- 5. Make Predictions ---
future_df['predicted_demand'] = model.predict(future_features).round().astype(int)

# Ensure predicted demand is within a reasonable range (e.g., 40-80 for 40 families)
future_df['predicted_demand'] = future_df['predicted_demand'].apply(lambda x: max(x, 40))
future_df['predicted_demand'] = future_df['predicted_demand'].apply(lambda x: min(x, 80))

print("\nPredicted Lunchbox Demand for Hari Raya Period:")
display_temp = future_df.copy()
display_temp['average_ingredient_cost_per_day'] = display_temp['average_ingredient_cost_per_day'].round(2) # Round to 2 decimal places
display_temp.insert(0, 'Day', range(1, 1 + len(display_temp)))
display(display_temp.set_index('Day'))
````

#### 4. Calculate Total Lunchbox Cost

Now, we will use the predicted lunchbox demand and the simulated `average_ingredient_cost_per_day` to estimate the total daily cost of providing the lunchboxes. For simplicity, let's assume a conversion factor from `average_ingredient_cost_per_day` to `cost_per_lunchbox` (e.g., 2.5 times the average ingredient cost accounts for all ingredients, packaging, and a small overhead).

````python
# Assume a multiplier to convert average ingredient cost to total cost per lunchbox
cost_multiplier = 2.5 # This factor would be determined by actual costing

future_df['estimated_cost_per_lunchbox_RM'] = future_df['average_ingredient_cost_per_day'] * cost_multiplier
future_df['total_daily_cost_RM'] = future_df['predicted_demand'] * future_df['estimated_cost_per_lunchbox_RM']

print("\nPredicted Demand and Estimated Costs for Hari Raya Period:")
display_temp = future_df[['date', 'predicted_demand', 'estimated_cost_per_lunchbox_RM', 'total_daily_cost_RM']].copy()
display_temp['estimated_cost_per_lunchbox_RM'] = display_temp['estimated_cost_per_lunchbox_RM'].round(2)
display_temp['total_daily_cost_RM'] = display_temp['total_daily_cost_RM'].round(2)
display_temp.insert(0, 'Day', range(1, 1 + len(display_temp)))
display(display_temp.set_index('Day'))

# Calculate total cost for the entire 30-day period
total_cost_30_days = future_df['total_daily_cost_RM'].sum()
print(f"\nTotal estimated cost for 30 days of lunchboxes: RM {total_cost_30_days:.2f}")
````
